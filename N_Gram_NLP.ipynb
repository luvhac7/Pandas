```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import ngrams
from collections import Counter
nltk.download('punkt')
nltk.download('punkt_tab')
txt="Downloading package punkt to /home/lab/nltk_data.."
tkn=word_tokenize(txt.lower())
uni=Counter(tkn)
bi=Counter(ngrams(tkn,2))
vs=len(uni)
def predict(prev):
    can={}
    for w in uni:
        prob=((bi[prev]+1)/(uni[prev]+vs))
        can[w]=prob
        return max(can,key=can.get)
w=input("Enter Word:").lower()
print("Next Word:",predict(w))
```
